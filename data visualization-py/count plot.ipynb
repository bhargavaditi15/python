# Count PLot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#load dataset
#load data 
var = sns.load_dataset("tips")
var
```

```python
sns.countplot(x="sex",data=var)
plt.show()
```

```python
sns.barplot(x="sex",y="size",data=var)
```

```python
sns.barplot(x="sex",y="size",data=var, hue="smoker")
```

```python
sns.barplot(y="sex",x="size",data=var, hue="smoker")
```

```python
sns.barplot(x="sex",y="size",data=var, hue="smoker",palette="bwr")
```

```python
sns.barplot(x="sex",y="size",data=var, hue="smoker",color="y")
```

```python
sns.barplot(x="sex",y="size",data=var, hue="smoker",color="r")
```

```python
sns.barplot(x="sex",y="size",data=var, hue="smoker",color="y",saturation=0.5)
```
