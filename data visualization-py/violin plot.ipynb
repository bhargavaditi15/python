# Violin Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#load dataset
#load data 
var = sns.load_dataset("tips")
var
```

```python
sns.violinplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue="time")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var,linewidth=2,palette="Dark2")
plt.show()
```

```python
sns.violinplot(x="time",y="total_bill",data=var,linewidth=2,palette="Dark2",order=["Dinner","Lunch"])
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var, hue="sex",split=True, sclae="width")
plt.show()
```

```python
sns.violinplot(x="total_bill",y="day",data=var, hue="sex",split=True, )
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=var, hue="sex",split=True, sclae="width",
              inner="quart")
plt.show()
```
