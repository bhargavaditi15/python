# Scatter Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#load data 
var = sns.load_dataset("penguins")
var
```

```python
sns.scatterplot(x="bill_length_mm",
               y="bill_depth_mm",data=var,hue="sex")

plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",
               y="bill_depth_mm",data=var,hue="sex",style="sex")

plt.show()
```

```python
#adjusting the size using sizes
sns.scatterplot(x="bill_length_mm",
               y="bill_depth_mm",data=var,hue="sex",style="sex",size="sex",sizes=(100,40))

plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",
               y="bill_depth_mm",data=var,hue="sex",style="sex",size="sex",sizes=(50,50),
               palette='gist_ncar')

plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",
               y="bill_depth_mm",data=var,hue="sex",style="sex",size="sex",sizes=(50,50),
               palette='gist_ncar', alpha=0.5)

plt.show()
```

```python
#change the marker
m = {"Male":"*","Female":"o"}
sns.scatterplot(x="bill_length_mm",
               y="bill_depth_mm",data=var,hue="sex",style="sex",size="sex",sizes=(50,50),
               palette='gist_ncar', alpha=1,markers = m)

plt.show()
```
