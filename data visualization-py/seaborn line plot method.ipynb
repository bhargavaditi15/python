```python
#load data 
y_1 = sns.load_dataset("penguins").head(20)
y_1
```

```python
#used to put lables
sns.lineplot(x="bill_depth_mm", y="flipper_length_mm",data=y_1)

#to remove the lable
plt.show()
```

```python
#kisi particular data ke respect mai chart bnabna hai to "hue" ka use hoga

#used to put lables
sns.lineplot(x="bill_depth_mm", y="flipper_length_mm",data=y_1,hue="sex")

#to remove the lable
plt.show()
```

```python
#customise the size


#used to put lables
sns.lineplot(x="bill_depth_mm", y="flipper_length_mm",data=y_1,hue="sex",size=50)

#to remove the lable
plt.show()
```

```python
#style  parameter


#used to put lables
sns.lineplot(x="bill_depth_mm", y="flipper_length_mm",data=y_1,hue="sex", style="sex")

#to remove the lable
plt.show()
```

```python
#palette ke through clour ko change kr sakte hai


#used to put lables
sns.lineplot(x="bill_depth_mm", y="flipper_length_mm",data=y_1,hue="sex", style="sex",palette="Accent")

#to remove the lable
plt.show()

#saare palette dekhne ke liye sirf koi letter likh ke run krdo saare palette ki list aajyegi
```

```python
#we want marker in the corner points
#used to put lables
sns.lineplot(x="bill_depth_mm", y="flipper_length_mm",data=y_1,hue="sex", style="sex",palette="Accent", markers=["o", ">"])

#to remove the lable
plt.show()
```

```python
#dotted line ko remove krne ke liye dash parameter use hoga

#used to put lables
sns.lineplot(x="bill_depth_mm", y="flipper_length_mm",data=y_1,hue="sex", style="sex",palette="Accent", markers=["o", ">"],
            dashes=False)

#to remove the lable
plt.show()
```

```python
#to remove legend we need to false it, legends ko laage ke liye function ki zarurat nahi h pr remove krne ke liye hai

#used to put lables
sns.lineplot(x="bill_depth_mm", y="flipper_length_mm",data=y_1,hue="sex", style="sex",palette="Accent", markers=["o", ">"],
            dashes=False, legend=False)

#legends ke parameter ko dekhne ke liye kuch bhi letter krke run krdo error mai pata chal jyega

#to remove the lable
plt.show()
```

```python
#title or grid ke liye

#used to put lables
sns.lineplot(x="bill_depth_mm", y="flipper_length_mm",data=y_1,hue="sex", style="sex",palette="Accent", markers=["o", ">"],
            dashes=False, legend=False)

#to remove the lable
plt.grid()
plt.title("python")
plt.show()
```
