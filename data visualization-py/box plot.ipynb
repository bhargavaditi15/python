# Box Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd
```

```python
#load dataset
var = sns.load_dataset("tips")
var
```

```python
sns.set(style="whitegrid")
#box ko kese plot kre
sns.boxplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.set(style="whitegrid")
#box ko kese plot kre
sns.boxplot(x="day",y="total_bill",data=var,hue="time")
plt.show()
```

```python
sns.set(style="whitegrid")
#box ko kese plot kre
sns.boxplot(x="day",y="total_bill",data=var,hue="time",color="g")
plt.show()
```

```python
sns.set(style="whitegrid")
#box ko kese plot kre
sns.boxplot(x="day",y="total_bill",data=var,hue="time",color="g",order=["Fri","Sun","Thur","Sat"])
plt.show()
```

```python
sns.set(style="whitegrid")
#box ko kese plot kre
sns.boxplot(x="day",y="total_bill",data=var,hue="time",color="r",order=["Fri","Sun","Thur","Sat"],
           showmeans=True)
plt.show()
```

```python
sns.set(style="whitegrid")
#box ko kese plot kre
#means property
sns.boxplot(x="day",y="total_bill",data=var,hue="time",color="r",order=["Fri","Sun","Thur","Sat"],
           showmeans=True,meanprops={"marker":"+","markeredgecolor":"b"})
plt.show()
```

```python
sns.set(style="whitegrid")
#box ko kese plot kre
#means property
sns.boxplot(x="day",y="total_bill",data=var,hue="time",color="r",order=["Fri","Sun","Thur","Sat"],
           showmeans=True,meanprops={"marker":"+","markeredgecolor":"b"},linewidth=3)
plt.show()
```

```python
sns.set(style="whitegrid")
#box ko kese plot kre
#means property
sns.boxplot(x="day",y="total_bill",data=var,hue="time",color="r",order=["Fri","Sun","Thur","Sat"],
           showmeans=True,meanprops={"marker":"+","markeredgecolor":"b"},linewidth=3,palette="RdPu")
plt.show()
```

```python
sns.set(style="whitegrid")
#box ko kese plot kre
#means property
sns.boxplot(x="day",y="total_bill",data=var,hue="time",color="r",order=["Fri","Sun","Thur","Sat"],
           showmeans=True,meanprops={"marker":"+","markeredgecolor":"b"},linewidth=3,palette="RdPu",
           orient="v")
plt.show()
```

```python
sns.set(style="whitegrid")
#box ko kese plot kre
#means property
sns.boxplot(x="total_bill",y="day",data=var,hue="time",color="r",order=["Fri","Sun","Thur","Sat"],
           showmeans=True,meanprops={"marker":"+","markeredgecolor":"b"},linewidth=3,palette="RdPu")
plt.show()
```

```python
sns.set(style="whitegrid")
#box ko kese plot kre
#means property
sns.boxplot(data=var,
           showmeans=True,meanprops={"marker":"+","markeredgecolor":"b"},linewidth=3,palette="RdPu",orient="h")
plt.show()
```

```python
sns.set(style="whitegrid")
#box ko kese plot kre
#means property
sns.boxplot(x=var["total_bill"])
plt.show()
```
