# Bar Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd
```

```python
#load dataset
var = sns.load_dataset("penguins")
var
```

```python
#sns.barplot(x=var.island,y=var.bill_length_mm)
sns.barplot(x="island",y="bill_length_mm",data=var)
#remove the line above 
plt.show()
```

```python
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex")
#remove the line above 
plt.show()
```

```python
#order parameter denge particular order set krne ke liye
order_1 = ["Dream","Torgersen","Biscore island"]
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",
           order = order_1)
plt.show()
```

```python
#female ko aage or male ko piche krne ke liye hue ka use krenge
order_1 = ["Dream","Torgersen","Biscore"]
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",
           order = order_1,hue_order=["Female","Male"])
plt.show()
```

```python
#multi bar graph bnane ke liye process use hota hai 

#kisi interval ko kam zyada krne ke liye ci use hoga
order_1 = ["Dream","Torgersen","Biscore"]
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",
           order = order_1,hue_order=["Female","Male"],ci=100)
plt.show()
```

```python
#n-boot size ke liye responsible hota hai
order_1 = ["Dream","Torgersen","Biscore"]
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",
           order = order_1,hue_order=["Female","Male"],ci=100,n_boot=2)
plt.show()
```

```python
#orientation isko horizontal krne ke liye ya vertical krne ke liye use krte hai
#horizontal mai error dega catagorical value nhi hona chaiye
order_1 = ["Dream","Torgersen","Biscore"]
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",
           order = order_1,hue_order=["Female","Male"],orient="v")
plt.show()
```

```python
#horizontal krne ka process

order_1 = ["Dream","Torgersen","Biscore"]
sns.barplot(x="bill_length_mm",y="bill_length_mm",data=var,orient="h")
plt.show()
```

```python
#color ko change krne ke liye color parameter ko use krenge 
order_1 = ["Dream","Torgersen","Biscore"]
sns.barplot(x="island",y="bill_length_mm",data=var, color="y")
plt.show()
```

```python
#palette parameter
order_1 = ["Dream","Torgersen","Biscore"]
sns.barplot(x="island",y="bill_length_mm",data=var, color="y",
           palette='icefire')
plt.show()
```

```python
#saturation color thoda blur ya change ho jyega
order_1 = ["Dream","Torgersen","Biscore"]
sns.barplot(x="island",y="bill_length_mm",data=var, color="y",
           palette='icefire',saturation=0.3)
plt.show()
```

```python
#error ka color ya width change krne ke liye
order_1 = ["Dream","Torgersen","Biscore"]
sns.barplot(x="island",y="bill_length_mm",data=var, color="y",
           palette='icefire',saturation=0.3,errcolor="r",errwidth=10)
plt.show()
```

```python
#cap size, iska use 

order_1 = ["Dream","Torgersen","Biscore"]
sns.barplot(x="island",y="bill_length_mm",data=var, color="y",
           palette='icefire',saturation=0.3,errcolor="r",capsize=0.5)
plt.show()
```

```python
#graph ko ek ke upper ek show krwane ke liye
order_1 = ["Dream","Torgersen","Biscore"]
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",
           order = order_1,hue_order=["Female","Male"],orient="v",dodge=False)
plt.show()
```

```python
#alpha parameter graph ko blur krne ke liye use hota hai
order_1 = ["Dream","Torgersen","Biscore"]
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",
           order = order_1,hue_order=["Female","Male"],orient="v",alpha=0.9)
plt.show()
```

```python
#to show the grid 
sns.set(style="darkgrid")
order_1 = ["Dream","Torgersen","Biscore"]
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",
           order = order_1,hue_order=["Female","Male"],orient="v",alpha=0.9)
plt.show()
```
