# PLot Styling

```python
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
#load dataset
var = sns.load_dataset("tips")
var
```

```python
sns.boxplot(y=var["total_bill"])
```

```python
sns.barplot(x="day",y="total_bill",data=var)
```

```python
sns.barplot(x="day",y="total_bill",data=var)
plt.grid()
plt.show()
```

```python
sns.set_style("white")
sns.barplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.set_style("dark")
sns.barplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.set_style("whitegrid")
sns.barplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.set_style("darkgrid")
sns.barplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.set_style("ticks")
sns.barplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.set_style("darkgrid")
sns.barplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
sns.set_style("whitegrid")
sns.barplot(x="day",y="total_bill",data=var)
sns.despine()
plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(12,12))
sns.barplot(x="day",y="total_bill",data=var)
sns.despine()
plt.show()
```

```python
sns.set_style("white")

sns.set_context("poster")

sns.barplot(x="day",y="total_bill",data=var)

plt.show()
```

```python
sns.set_style("white")

sns.set_context("notebook")

sns.barplot(x="day",y="total_bill",data=var)

plt.show()
```

```python
sns.set_style("white")

sns.set_context("poster", font_scale=2)

sns.barplot(x="day",y="total_bill",data=var)

plt.show()
```

```python
sns.set_style("white")

sns.set_context("talk", font_scale=2)

sns.barplot(x="day",y="total_bill",data=var)

plt.show()
```

```python
sns.set_style("white")

sns.set_context("talk", font_scale=2)

sns.barplot(x="day",y="total_bill",data=var,palette="cool_r")

plt.show()
```
