# melt()

```python
import pandas as pd
```

```python
var = pd.DataFrame({"days":[1,2,3,4,5,6], "eng":[10,12,14,15,16,12], "maths":[17,18,19,13,14,16]})

var
```

```python
pd.melt(var)
```

```python
pd.melt(var, id_vars=["eng"])
```

```python
pd.melt(var, id_vars=["days"])
```

```python
pd.melt(var, id_vars=["eng"], var_name="python")
```

```python
pd.melt(var, id_vars=["eng"], var_name="python", value_name="wscube")
```

# pivot()

```python
var = pd.DataFrame({"days":[1,2,3,4,5,6], 
                    "st_name":["a","b","c","a","b","c"],
                    "eng":[10,12,14,15,16,12],
                    "maths":[17,18,19,13,14,16]})

var
```

```python
var.pivot(index="days", columns="st_name")
```

```python
var = pd.DataFrame({"days":[1,2,3,4,5,6], 
                    "st_name":["a","b","a","a","b","b"],
                    "eng":[10,12,14,15,16,12],
                    "maths":[17,18,19,13,14,16]})

var
```

```python
var.pivot(index="days", columns="st_name")
```

```python
var.pivot(index="days", columns="st_name", values="eng")
```

```python
var = pd.DataFrame({"days":[1,1,1,1,2,2], 
                    "st_name":["a","b","a","a","b","b"],
                    "eng":[10,12,14,15,16,12],
                    "maths":[17,18,19,13,14,16]})

var
```

```python
var.pivot_table(index="st_name", columns="days",
               aggfunc="mean")
```

```python
var.pivot_table(index="st_name", columns="days",
               aggfunc="sum")
```

```python
var.pivot_table(index="st_name", columns="days",
               aggfunc="mean", margins="True")
```
