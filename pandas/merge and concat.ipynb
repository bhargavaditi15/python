# merge 

```python
import pandas as pd
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2 = pd.DataFrame({"A":[1,2,3,4],"C":[21,22,23,24]})
```

```python
pd.merge(var1,var2,on="A")
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2 = pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})
```

```python
pd.merge(var1,var2,on="A")
```

```python
pd.merge(var1,var2, how="inner")
```

```python
pd.merge(var1,var2, how="left")
```

```python
pd.merge(var1,var2, how="right")
```

```python
pd.merge(var1,var2, how="outer")
```

```python
pd.merge(var1,var2, how="outer",indicator = True)
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2 = pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})
```

```python
pd.merge(var1,var2,left_index=True, right_index=True,)
```

```python
pd.merge(var1,var2,left_index=True, right_index=True, suffixes=("name","python"))
```

# concat

```python
sr1 = pd.Series([1,2,3,4])
sr2 = pd.Series([11,21,31,41])
```

```python
pd.concat([sr1,sr2])
```

```python
d1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2 = pd.DataFrame({"A":[1,2,3,4],"C":[21,22,23,24]})
```

```python
pd.concat([d1,d2])
```

```python
pd.concat([d1,d2],axis=1)
```

```python
a1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
a2 = pd.DataFrame({"A":[1,2],"C":[21,22]})

pd.concat([a1,a2],axis=1, join="outer")
```

```python
pd.concat([a1,a2],axis=1, join="inner")
```

```python
a1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
a2 = pd.DataFrame({"A":[1,2,3,4],"C":[21,22,23,24]})

pd.concat([a1,a2],axis=1, keys=["a1","a2"])
```

```python
a1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
a2 = pd.DataFrame({"A":[1,2,3,4],"C":[21,22,23,24]})

pd.concat([a1,a2],axis=0, keys=["a1","a2"])
```

```python
a1 = pd.DataFrame({"A":[1,2,3,4]})
a2 = pd.DataFrame({"A":[1,2,3,4],"C":[21,22,23,24]})

pd.concat([a1,a2])
```

```python
a1 = pd.DataFrame({"D":[1,2,3,4]})
a2 = pd.DataFrame({"A":[1,2,3,4],"C":[21,22,23,24]})

pd.concat([a1,a2])
```
