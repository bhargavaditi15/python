# handling missing value (dropna & fillna)

```python
import pandas as pd
```

```python
csv = pd.read_csv("HR-dataset.csv")
```

```python
csv
```

```python
#drop missing data
csv.dropna()
```

```python
#column ke along drop krna
# 1 se coloum hota hai 0 dene pe row along hota hai
csv.dropna(axis=1)
```

```python
csv.dropna(axis=0)
```

```python
csv.dropna(how="any")#puri empty row ko hata deta hai
```

```python
csv = pd.read_csv("HR-dataset.csv")
```

```python
csv
```

```python
csv.dropna(subset=["EmpID"]) #particular colum ki null value chali jaati hai
```

```python
csv.dropna(inplace=True)
csv
```

```python
#ek column mai null value hai pr sirf ek he null value wale ko drop krna hai
#csv.drop(thresh = 1)
```

```python
csv = pd.read_csv("HR-dataset.csv")
csv
csv.fillna("python")
```

```python
csv.fillna({"Age":"aditi","Attrition":"gaurja","DistanceFromHome":"sona"})
```

```python
#forward filling
csv.fillna(method="ffill")
```

```python
#backward filling
csv.fillna(method="bfill")
```

```python
#axis ke along fill hona
csv.fillna(method="ffill",axis=0)
```

```python
csv.fillna(12,inplace=True)
```

```python
csv
```

```python
csv.fillna("python",limit =2)
```
